# Plot constraints on $C_7$ and $C_7'$

This notebook generates plots that show the constraints on the Wilson coeffiencts $C_7$ and $C_7'$.

It is based on a [notebook](https://github.com/DavidMStraub/paper-bvgamma-ps) provided as compagnion script to the paper [arXiv:1608.02556](https://arxiv.org/abs/1608.02556) by Ayan Paul and David M. Straub.

It requires [flavio](https://flav-io.github.io/) v0.21.3 or later.

## Preliminaries

Loading packages

```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
import pickle
import flavio
import flavio.plots
import matplotlib.pyplot as plt
```

Loading data

```python
with open('./data.p', 'rb') as f:
    (data,fits) = pickle.load(f)
```

## Results

```python
labels = {
    'global': 'global',
    'BR': 'branching ratios',
    'A': flavio.Observable.get_instance(fits['A'][0]).tex,
    'S': flavio.Observable.get_instance(fits['S'][0]).tex,
    'ATIm': flavio.Observable.get_instance(fits['ATIm'][0][0]).tex,
    'P1': flavio.Observable.get_instance(fits['P1'][0][0]).tex,
    'no_BR': 'non-BR',
    'BR(B0->K*gamma)' : r'$\mathcal{B}(B^0\to K^*\gamma)$',
    'BR(B+->K*gamma)' : r'$\mathcal{B}(B^+\to K^*\gamma)$',
    'BR(B->Xsgamma)' : r'$\mathcal{B}(B\to X_s\gamma)$',
    'BR(Bs->phigamma)' : r'$\mathcal{B}(B_s\to \phi\gamma)$',
}
```

```python
fig=plt.figure(figsize=(5,5))
plt.xlim([-0.2,1.0])
plt.ylim([-0.6,0.6])

plot_list = [
    ['A',2],
    ['P1',3],
    ['S',4],
    ['BR',1],
]
for f,i in plot_list:
    flavio.plots.contour(interpolation_factor=10,
                             col=i,
                             label=labels[f],
                             **data[f])
flavio.plots.contour(interpolation_factor=10,
                         col=0,
                         label=labels['global'],
                         **data['global'])
flavio.plots.flavio_branding(x=0.05, y=0.09)
plt.axhline(0, c='k', lw=0.2)
plt.axvline(0, c='k', lw=0.2)
plt.xlabel(r'$\text{Re}(C_7^\text{NP})$')
plt.ylabel(r'$\text{Re}(C_7^{\prime\,\text{NP}})$')
plt.xlim([-0.2,1.0])
plt.ylim([-0.6,0.6])
plt.legend(loc=1, bbox_to_anchor=(1, 1))
plt.tight_layout()
#plt.savefig('C7_C7p.pdf')
```

```python
fig=plt.figure(figsize=(5,5))
plt.xlim([-0.2,1.0])
plt.ylim([-0.6,0.6])

plot_list = [
    ['no_BR',3],
    ['BR(B0->K*gamma)',1],
    ['BR(B+->K*gamma)',5],
    ['BR(Bs->phigamma)',2],
    ['BR(B->Xsgamma)',4],
]
for f,i in plot_list:
    flavio.plots.contour(interpolation_factor=10,
                             col=i,
                             label=labels[f],
                             **data[f])
flavio.plots.contour(interpolation_factor=10,
                         col=0,
                         label=labels['global'],
                         **data['global'])
flavio.plots.flavio_branding(x=0.05, y=0.09)
plt.axhline(0, c='k', lw=0.2)
plt.axvline(0, c='k', lw=0.2)
plt.xlabel(r'$\text{Re}(C_7^\text{NP})$')
plt.ylabel(r'$\text{Re}(C_7^{\prime\,\text{NP}})$')
plt.xlim([-0.2,1.0])
plt.ylim([-0.6,0.6])
plt.legend(loc=1, bbox_to_anchor=(1, 1))
plt.tight_layout()
#plt.savefig('C7_C7p_BR_separately.pdf')
```
